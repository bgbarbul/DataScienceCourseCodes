```python
stat = df.groupby(["visit_date","Hétnap"],as_index=False).agg({"visit_spend":"sum"})
```

```python
#11:45
```

```python
stat
```

```python
stat.columns= ['visit_date', 'hetnapja', 'napi_koltes']
```

```python
stat
```

```python
heti_statisztika = stat.groupby('hetnapja', as_index =False).agg({'napi_koltes' : 'mean'})
```

```python
heti_statisztika
```

```python
heti_statisztika.columns
```

```python
heti_statisztika.columns= ['hetnapja', 'tipikus_napi_koltes']
```

```python
heti_statisztika
```

```python
egyben = stat.merge(heti_statisztika,on='hetnapja', how='left') #join
```

```python
egyben
```

```python
egyben['furcsasag_index']=egyben['napi_koltes']/egyben['tipikus_napi_koltes']
```

```python
egyben['furasag_mertek'] = abs(1-egyben['furcsasag_index'])
```

```python
egyben.sort_values('furasag_mertek',ascending=False).head(20)
```
